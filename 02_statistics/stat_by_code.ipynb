```python
import numpy as np
import pandas as pd

df = pd.read_csv("orange_cap.csv")
df.columns
```

```python
runs = df['Runs']
```

```python
mean_runs = runs.mean()
median_runs = runs.median()
mode_runs = runs.mode().tolist()

range_runs = runs.max()- runs.min()
std_runs = runs.std()
var_runs = runs.var()

q1 = runs.quantile(0.25)
q2 = runs.quantile(0.55)
q3 = runs.quantile(0.75)
iqr_runs = q3-q2

(mean_runs, median_runs, mode_runs, range_runs, std_runs, var_runs, q1, q2, q3, iqr_runs)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.boxplot(x=df['Runs'], color='skyblue')
plt.title('boxplot of runs')
plt.xlabel('Runs')

plt.subplot(1,2,2)
sns.histplot(df['Runs'], kde=True, color="orange", bins=10)
plt.title("Dist of runs")
plt.xlabel("Runs")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
