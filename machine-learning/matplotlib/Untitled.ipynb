## Matplotlib

### ðŸŽ¨ 1. **Figure and Axes Management**

*Think: Where am I painting?*

| Purpose            | Function         | Example                          |
| ------------------ | ---------------- | -------------------------------- |
| Create canvas      | `plt.figure()`   | `plt.figure(figsize=(8,5))`      |
| Create subplots    | `plt.subplots()` | `fig, ax = plt.subplots(2,1)`    |
| Get current figure | `plt.gcf()`      | `plt.gcf().set_size_inches(6,4)` |
| Get current axes   | `plt.gca()`      | `plt.gca().set_title("Title")`   |

---

### ðŸ“Š 2. **Plotting Basics**

*Think: What shapes do I show?*

| Purpose      | Function        | Example                         |
| ------------ | --------------- | ------------------------------- |
| Line chart   | `plt.plot()`    | `plt.plot(x, y)`                |
| Bar chart    | `plt.bar()`     | `plt.bar(categories, values)`   |
| Scatter plot | `plt.scatter()` | `plt.scatter(x, y)`             |
| Histogram    | `plt.hist()`    | `plt.hist(data, bins=10)`       |
| Pie chart    | `plt.pie()`     | `plt.pie(sizes, labels=labels)` |

---

### ðŸŽ¯ 3. **Axis and Labels**

*Think: How does the viewer read it?*

| Purpose         | Function                        | Example                          |
| --------------- | ------------------------------- | -------------------------------- |
| X/Y axis label  | `plt.xlabel()` / `plt.ylabel()` | `plt.xlabel("Time")`             |
| Title           | `plt.title()`                   | `plt.title("Revenue Over Time")` |
| X/Y ticks       | `plt.xticks()` / `plt.yticks()` | `plt.xticks(rotation=45)`        |
| Set axis limits | `plt.xlim()` / `plt.ylim()`     | `plt.xlim(0, 100)`               |

---

### ðŸ§± 4. **Grid, Legend, Text, and Annotation**

*Think: Context and clarity.*

| Purpose              | Function         | Example                                           |
| -------------------- | ---------------- | ------------------------------------------------- |
| Add grid             | `plt.grid()`     | `plt.grid(True)`                                  |
| Add legend           | `plt.legend()`   | `plt.legend(['Model A'])`                         |
| Add text inside plot | `plt.text()`     | `plt.text(2, 10, 'Peak')`                         |
| Annotate a point     | `plt.annotate()` | `plt.annotate("Max", xy=(x,y), xytext=(x+1,y+2))` |

### ðŸ§¤ 5. **Styling & Customization**

*Think: Colors, shapes, textures.*

## 1. Figure and Axes Management

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(2,1,1) # nrow, ncols, index

ax.plot([1,2,3],[10,20,55])
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot([1, 2, 3], [1, 4, 9], label='x squared')
ax.plot([1, 2, 3], [1, 2, 3], label='x linear')
ax.plot([1, 2, 3], [2, 4, 6], label='x linear')
ax.set_title("Multiple Lines")
ax.legend()

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

axes[0, 0].plot([1, 2, 3], [1, 2, 3])
axes[0, 1].bar([1, 2, 3], [3, 2, 1])
axes[1, 0].scatter([1, 2, 3], [3, 2, 1])
axes[1, 1].hist([1, 1, 2, 3, 3, 3])

fig.suptitle("Different Plots in Grid")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4), facecolor='lightgrey')

ax.plot([1, 2, 3], [10, 5, 0])
ax.set_title("Wider Canvas")

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

ax1.plot([1, 2, 3], [3, 2, 1])
ax2.plot([1, 2, 3], [1, 2, 3])

ax1.set_title("Shared X-Axis")
plt.show()
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()  # create a second y-axis

ax1.plot([1, 2, 3], [1, 2, 3], 'g-')
ax2.plot([1, 2, 3], [10, 100, 1000], 'b-')

ax1.set_ylabel('Linear Scale', color='g')
ax2.set_ylabel('Log Scale', color='b')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)  # 1 row, 1 col, 1st plot

ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("Basic Line Plot")

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [10, 20, 30])
ax.set_title("Subplots Syntax")

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(6, 6))

axs[0, 0].plot([1, 2], [1, 2])
axs[0, 1].bar([1, 2], [3, 1])
axs[1, 0].scatter([1, 2], [2, 1])
axs[1, 1].hist([1, 1, 2, 3, 3, 3])

fig.suptitle("2x2 Layout of Axes")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(6, 4))
gs = gridspec.GridSpec(2, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, :])  # top row
ax2 = fig.add_subplot(gs[1, 0])  # bottom-left
ax3 = fig.add_subplot(gs[1, 1])  # bottom-right

ax1.plot([1, 2], [1, 2])
ax2.bar([1, 2], [3, 1])
ax3.hist([1, 1, 2, 3])

fig.suptitle("GridSpec Layout")
plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot([1, 2, 3], [10, 20, 30], 'g-')
ax2.plot([1, 2, 3], [1000, 2000, 3000], 'b-')

ax1.set_ylabel("Green Axis", color='g')
ax2.set_ylabel("Blue Axis", color='b')

plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(5, 8))

axs[0].plot([1, 2, 3], [1, 4, 9])
axs[1].plot([1, 2, 3], [9, 4, 1])
axs[2].plot([1, 2, 3], [1, 2, 1])

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axs[i, j].plot([1, 2, 3], [1*(i+j), 2*(i+j), 3*(i+j)])

fig.suptitle("Shared Axes in 2x2")

plt.show()
```

```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([2.5, 3.5, 4.0])
error = np.array([0.2, 0.3, 0.1])

plt.errorbar(x, y, yerr=error, fmt='o', color='black', capsize=5)
plt.title("Error Bar Plot")
plt.show()
```

```python
train_acc = [0.5, 0.6, 0.7, 0.75, 0.78]
val_acc = [0.45, 0.55, 0.65, 0.7, 0.65]

plt.plot(train_acc, label='Train Acc')
plt.plot(val_acc, label='Validation Acc')
plt.title("Learning Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_redundant=0, random_state=42)
model = LogisticRegression().fit(X, y)

# Plot
xx, yy = np.meshgrid(np.linspace(X[:,0].min(), X[:,0].max(), 100),
                     np.linspace(X[:,1].min(), X[:,1].max(), 100))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='bwr')
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr', edgecolors='k')
plt.title("Logistic Regression Decision Boundary")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('sin(x)')

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('cos(x)')

axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title('tan(x)')

axs[1, 1].plot(x, np.exp(x / 5))
axs[1, 1].set_title('exp(x)')

fig.suptitle('Multiple Functions')
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, sharex=True)
x = np.linspace(0, 10, 100)

axs[0].plot(x, np.sin(x), label="sin")
axs[1].plot(x, np.cos(x), label="cos", color='orange')

axs[0].legend()
axs[1].legend()
fig.suptitle("Shared X-Axis")
plt.show()
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(3, 3)

ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1:, 0])
ax3 = fig.add_subplot(gs[1:, 1:])
ax1.set_title("Top Row")
ax2.set_title("Left Column")
ax3.set_title("Bottom Right")
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 6))
gs = fig.add_gridspec(2, 2)

# Main plots
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])

# Nested inside bottom-right
sub_gs = gs[1, :].subgridspec(1, 2)
ax3 = fig.add_subplot(sub_gs[0, 0])
ax4 = fig.add_subplot(sub_gs[0, 1])

ax1.set_title("Top-Left")
ax2.set_title("Top-Right")
ax3.set_title("Bottom-Left")
ax4.set_title("Bottom-Right")

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)

# Add inset
ax_inset = inset_axes(ax, width="30%", height="30%", loc='upper right')
ax_inset.plot(x, y)
ax_inset.set_xlim(2, 4)
ax_inset.set_ylim(0, 1)

plt.title("Inset View")
plt.show()
```

```python
x = np.arange(0, 10)
y1 = np.exp(x)
y2 = np.log1p(x)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-', label='exp')
ax2.plot(x, y2, 'b-', label='log')

ax1.set_ylabel('exp(x)', color='g')
ax2.set_ylabel('log(1+x)', color='b')
plt.title("Two Y-Axes")
plt.show()
```
