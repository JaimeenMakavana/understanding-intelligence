## ðŸŽ¨ Seaborn: Categorized by Functionality
---

### **1. Distribution Plots**

> *Reveal the nature of a single variable.*

| Plot Type | Function         | Use                       |
| --------- | ---------------- | ------------------------- |
| Histogram | `sns.histplot()` | Shape, skewness, bin size |
| KDE Plot  | `sns.kdeplot()`  | Smooth probability curve  |
| ECDF Plot | `sns.ecdfplot()` | Cumulative distribution   |
| Rug Plot  | `sns.rugplot()`  | Raw data ticks on axis    |

---

### **2. Relational Plots**

> *Relationship between two or more variables.*

| Plot Type       | Function            | Use                        |
| --------------- | ------------------- | -------------------------- |
| Scatter Plot    | `sns.scatterplot()` | Trend, clusters            |
| Line Plot       | `sns.lineplot()`    | Continuous trend           |
| Relational Grid | `sns.relplot()`     | Faceted scatter/line plots |

---

### **3. Categorical Plots**

> *Visualize data across categories.*

| Plot Type   | Function           | Use                         |
| ----------- | ------------------ | --------------------------- |
| Strip Plot  | `sns.stripplot()`  | Raw value scatter           |
| Swarm Plot  | `sns.swarmplot()`  | Like strip, no overlap      |
| Box Plot    | `sns.boxplot()`    | Median, quartiles, outliers |
| Violin Plot | `sns.violinplot()` | KDE + boxplot hybrid        |
| Bar Plot    | `sns.barplot()`    | Mean with CI                |
| Count Plot  | `sns.countplot()`  | Frequency of categories     |
| Point Plot  | `sns.pointplot()`  | Line across categories      |

---

### **4. Matrix Plots**

> *Show patterns across grids of values.*

| Plot Type  | Function           | Use                           |
| ---------- | ------------------ | ----------------------------- |
| Heatmap    | `sns.heatmap()`    | Correlation matrix, confusion |
| Clustermap | `sns.clustermap()` | Dendrogram + heatmap          |

---

### **5. Regression & Model Plots**

> *Understand trends + relationships statistically.*

| Plot Type           | Function          | Use                          |
| ------------------- | ----------------- | ---------------------------- |
| Regression          | `sns.regplot()`   | Fit line + scatter           |
| Logistic Regression | `sns.lmplot()`    | Probabilistic classification |
| Residual Plot       | `sns.residplot()` | Residual error               |

---

### **6. Multi-Plot Grids (Facets)**

> *Small multiples â€” break data across dimensions.*

| Plot Type  | Function          | Use                        |
| ---------- | ----------------- | -------------------------- |
| Facet Grid | `sns.FacetGrid()` | Multi-panel by row/col/hue |
| Pair Plot  | `sns.pairplot()`  | All-by-all scatter matrix  |
| Joint Plot | `sns.jointplot()` | X-Y + marginals            |
| Cat Plot   | `sns.catplot()`   | Categorical + facet combo  |

---

## ðŸ”§ Also Important: Configuration & Themes

* `sns.set_theme()`: Set global styles
* `sns.set_context()`: Control font & spacing (talk, paper, notebook)
* `palette`, `style`, `hue`: Aesthetics by group
* `ax.set()`, `plt.figure()`, `tight_layout()`: Borrowed from matplotlib



# 1. Distribution Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set theme
sns.set_theme(style="whitegrid")
# Dummy data
data = np.random.normal(loc=0, scale=1, size=1000)
df = pd.DataFrame({'value': data})
print(df)
```

```python
# 1. Simple histogram
sns.histplot(data, bins=20)
plt.show()
```

```python
# 2. Histogram with KDE
sns.histplot(data, kde=True)
plt.show()
```

```python
# 3. KDE only
sns.kdeplot(data)
plt.show()
```

```python
# 4. KDE with fill
sns.kdeplot(data, fill=True)
plt.show()
```

```python
# 5. KDE with rugplot
sns.kdeplot(data)
sns.rugplot(data)
plt.show()
```

```python
# 6. Custom color & line style
sns.kdeplot(data, color='purple', linestyle='--')
plt.show()
```

```python
# 7. KDE with bandwidth control
sns.kdeplot(data, bw_adjust=0.5)
plt.show()
```

```python
# 8. Histogram with cumulative=True
sns.histplot(data, cumulative=True, kde=True)
plt.show()

```

```python
# 9. KDE cumulative
sns.kdeplot(data, cumulative=True)
plt.show()
```

```python
# 10. Rugplot alone
sns.rugplot(data, height=0.1)
plt.show()
```

```python
# 11. Grouped KDEs by category
df['group'] = np.where(np.random.rand(1000) > 0.5, 'A', 'B')
sns.kdeplot(data=df, x="value", hue="group")
plt.show()
```

```python
# 12. Fill KDEs by group
sns.kdeplot(data=df, x="value", hue="group", fill=True)
plt.show()
```

```python
# 13. Grouped histograms
sns.histplot(data=df, x="value", hue="group", bins=30)
plt.show()
```

```python
# 14. Grouped histograms (stacked)
sns.histplot(data=df, x="value", hue="group", multiple="stack")
plt.show()
```

```python
# 15. Grouped histograms (dodge)
sns.histplot(data=df, x="value", hue="group", multiple="dodge", bins=20)
plt.show()
```

```python
# 16. Multiple KDEs with different line styles
sns.kdeplot(data=df, x="value", hue="group", linestyle="--", common_norm=False)
plt.show()
```

```python
# 17. KDE with shading and alpha
sns.kdeplot(data=df, x="value", hue="group", fill=True, alpha=0.4)
plt.show()
```

```python

# 18. Histogram with edge coloring
sns.histplot(data, edgecolor="black", linewidth=1)
plt.show()
```

```python
# 19. KDE on log-scaled data
df_log = pd.DataFrame({'value': np.random.lognormal(mean=1.0, sigma=0.5, size=1000)})
sns.kdeplot(data=df_log, x="value")
plt.xscale("log")
plt.show()
```

```python
# 20. Overlaid KDE and histogram with different styles
sns.histplot(data, bins=30, color="lightgrey")
sns.kdeplot(data, color="red", lw=2)
plt.show()
```

```python
# 21. Faceted KDEs by group
g = sns.FacetGrid(df, col="group")
g.map(sns.kdeplot, "value")
plt.show()
```

```python
# 22. KDE with clipped range
sns.kdeplot(data, clip=(-2, 2))
plt.show()
```

```python
# 23. KDE of two different variables
df['value2'] = df['value'] + np.random.normal(scale=0.3, size=1000)
sns.kdeplot(df['value'], label='value')
sns.kdeplot(df['value2'], label='value2')
plt.legend()
plt.show()
```

```python
# 24. Multivariate KDE (2D)
sns.kdeplot(x=df['value'], y=df['value2'], fill=True)
plt.show()
```

```python
# 25. Contour + filled KDE
sns.kdeplot(x=df['value'], y=df['value2'], cmap="Blues", fill=True, thresh=0.05)
plt.show()
```

```python
# 26. KDE with vertical orientation
sns.kdeplot(data, vertical=True)
plt.show()
```

```python
# 27. KDE with shading under curve
sns.kdeplot(data, fill=True, color="orange")
plt.axhline(0, color='black')
plt.show()

```

```python
# 28. ECDF plot (cumulative distribution)
sns.ecdfplot(data)
plt.show()
```

```python
# 29. ECDF grouped
sns.ecdfplot(data=df, x="value", hue="group")
plt.show()
```

```python

# 30. KDE with markers at specific quantiles
sns.kdeplot(data, fill=True)
for q in [0.25, 0.5, 0.75]:
    plt.axvline(np.quantile(data, q), color='r', linestyle='--')
plt.show()
```

## Relational Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set theme
sns.set_theme(style="whitegrid")

# Base data
np.random.seed(42)
df = pd.DataFrame({
    "x": np.random.rand(100),
    "y": np.random.rand(100) + np.linspace(0, 1, 100),
    "category": np.random.choice(["A", "B"], size=100),
    "size": np.random.randint(10, 100, size=100),
    "time": np.linspace(1, 100, 100)
})
```

```python
df
```

```python
# 1. Basic scatter plot
sns.scatterplot(data=df, x="x", y="y")
plt.show()
```

```python
# 2. Colored by category
sns.scatterplot(data=df, x="x", y="y", hue="category")
plt.show()
```

```python
# 3. Scatter with size encoding
sns.scatterplot(data=df, x="x", y="y", size="size")
plt.show()
```

```python
# 4. Scatter with both hue and size
sns.scatterplot(data=df, x="x", y="y", hue="category", size="size")
plt.show()
```

```python
# 5. Line plot
sns.lineplot(data=df, x="x", y="y")
plt.show()
```

```python
# 6. Line plot with multiple lines (grouped)
sns.lineplot(data=df, x="time", y="y", hue="category")
plt.show()
```

```python
# 7. Scatterplot with markers
sns.scatterplot(data=df, x="x", y="y", style="category")
plt.show()
```

```python
# 8. Custom markers and colors
sns.scatterplot(data=df, x="x", y="y", hue="category", style="category", palette="coolwarm")
plt.show()
```

```python
# 9. Transparent points (alpha)
sns.scatterplot(data=df, x="x", y="y", alpha=0.6)
plt.show()
```

```python
# 10. Sorted lineplot
df_sorted = df.sort_values(by="time")
sns.lineplot(data=df_sorted, x="time", y="y")
plt.show()
```

```python
# 11. Style different groups by line style
sns.lineplot(data=df, x="time", y="y", hue="category", style="category")
plt.show()
```

```python
# 12. Highlight group differences with palette
sns.scatterplot(data=df, x="x", y="y", hue="category", palette="Set2")
plt.show()
```

```python
# 18. Faceted scatter plot by category
g = sns.relplot(data=df, x="x", y="y", col="category", kind="scatter")
g.set_titles("Category: {col_name}")
plt.show()
```

```python
# 21. Scatterplot with trend line (from regplot instead)
sns.regplot(x="x", y="y", data=df, scatter=True, ci=95)
plt.show()
```

```python
# 22. Binned scatterplot (hex)
sns.jointplot(data=df, x="x", y="y", kind="hex")
plt.show()

```

```python
# 23. 3D-like encoding: hue, size, style
sns.scatterplot(data=df, x="x", y="y", hue="category", size="size", style="category")
plt.show()

```

```python
# 24. Temporal animation idea (for real use: Plotly or matplotlib.animation)
for i in range(0, 100, 20):
    sns.scatterplot(data=df[df["time"] < i], x="x", y="y")
    plt.title(f"Time < {i}")
    plt.show()
```

```python
# 25. FacetGrid with multiple scatter facets
g = sns.FacetGrid(df, col="category")
g.map_dataframe(sns.scatterplot, x="x", y="y")
plt.show()
```

```python
# 26. Highlight outliers
df["outlier"] = df["y"] > 1.5
sns.scatterplot(data=df, x="x", y="y", hue="outlier")
plt.show()
```

```python
# 27. Add regression line to scatter
sns.lmplot(data=df, x="x", y="y", hue="category", markers=["o", "s"])
plt.show()
```

```python
# 28. Line plot with style and palette
sns.lineplot(data=df, x="time", y="y", hue="category", style="category", palette="Set1")
plt.show()
```

```python
# 30. PairGrid-style relational map
g = sns.PairGrid(df[["x", "y", "size"]])
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot)
plt.show()
```
