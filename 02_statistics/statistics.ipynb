scenarioes for hypothesis testing for feature selection




| Evaluation Task                      | Hypothesis Test         |
| ------------------------------------ | ----------------------- |
| CV Accuracy Comparison (same splits) | Paired t-test           |
| Accuracy (diff datasets)             | Independent t-test      |
| Proportions (accuracy / CTR)         | Z-test for proportions  |
| F1, Recall, Precision                | Wilcoxon / Mann-Whitney |
| ROC AUC Curves                       | DeLong Test             |
| Prod A/B test                        | Z-test / Chi-square     |
| Drift Detection                      | KS-test / t-test / Chi² |
